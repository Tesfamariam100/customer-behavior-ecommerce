```python
import pandas as pd

# Load dataset
file_path = "../ecommerce_customer_data_custom_ratios.csv"  # Adjust path if needed
df = pd.read_csv(file_path)

# Display basic info
df.info()
df.head()
```

```python
# Check for missing values
df.isnull().sum()

# Check for duplicate rows
df.duplicated().sum()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Age distribution
plt.figure(figsize=(8,5))
sns.histplot(df["Customer Age"], bins=30, kde=True, color="skyblue")
plt.title("Customer Age Distribution")
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()

# Gender distribution
df["Gender"].value_counts().plot(kind="bar", color=["blue", "pink"])
plt.title("Customer Gender Distribution")
plt.xlabel("Gender")
plt.ylabel("Count")
plt.show()
```

```python
# Most purchased product categories
plt.figure(figsize=(10,5))
df["Product Category"].value_counts().plot(kind="bar", color="gray")
plt.title("Most Popular Product Categories")
plt.xlabel("Category")
plt.ylabel("Number of Purchases")
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(8,5))
df["Payment Method"].value_counts().plot(kind="pie", autopct="%1.1f%%", colors=["gold", "lightblue", "green"])
plt.title("Preferred Payment Methods")
plt.ylabel("")
plt.show()
```

```python
import matplotlib.pyplot as plt

# Create figure
plt.figure(figsize=(6,5))
df["Churn"].value_counts().plot(kind="bar", color=["red", "green"])

# Add labels & title
plt.title("Churned vs Retained Customers")
plt.xlabel("Churn Status (0 = Retained, 1 = Churned)")
plt.ylabel("Count")

# Save the figure as a PNG file
plt.savefig("../reports/churn_analysis.png", dpi=300, bbox_inches="tight")  

# Show the plot
plt.show()
```
